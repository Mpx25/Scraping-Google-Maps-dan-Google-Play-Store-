```python
#Library yang dibutuhkan
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
import time, pandas as pd
```

```python
# Argumen agar menerima review dengan Bahasa Indonesia
options = Options()
options.add_argument('--lang=id')
options.add_argument('accept-languange=id-ID, id')
driver = webdriver.Chrome(options=options)

# Panggil lokasi digmaps lewat URL
driver.get("https://www.google.com/maps/place/Universitas+Ahmad+Dahlan+-+Kampus+4/@-7.833377,110.3805664,17z/data=!4m8!3m7!1s0x2e7a5701a2ae1c23:0x173dbeeddc56d9e!8m2!3d-7.833377!4d110.3831413!9m1!1b1!16s%2Fg%2F11b88tzs_l?entry=ttu&g_ep=EgoyMDI1MTIwOS4wIKXMDSoASAFQAw%3D%3D")
time.sleep(3)

# Cek apakah lokasi benar-benar ada
try:
    nama_tempat = driver.find_element(By.CLASS_NAME, "DUwDvf").text.strip()
except NoSuchElementException:
    nama_tempat = "unknown"

# Klik tombol More Reviews jika ada
try:
    more_reviews = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//span[contains(text(), 'Ulasan lainnya') or contains(text(), 'More reviews')]/ancestor::button")))
    more_reviews.click()
    print("[INFO] Klik tombol More reviews.")
    time.sleep(3)
except:
    print("[INFO] Tidak ada tombol More reviews.")

# Scraping data di Gmaps
reviews, seen = [], set()
while len(reviews) < 50:
    for c in driver.find_elements(By.CSS_SELECTOR, 'div.jftiEf'):
        user_el = c.find_elements(By.CLASS_NAME, 'd4r55')
        review_el = c.find_elements(By.CLASS_NAME, 'wiI7pd')
        rating_el = c.find_elements(By.CLASS_NAME, 'kvMYJc')
        # Skip jika review kosong
        if not review_el or not review_el[0].text.strip():
            continue
        user = user_el[0].text if user_el else 'Unknown'
        review = review_el[0].text.strip()
        rating = rating_el[0].get_attribute('aria-label').split()[0] if rating_el else 'Unknown'
        if (user, review) not in seen:
            seen.add((user, review))
            reviews.append({
                'nama_tempat': nama_tempat,
                'user': user,
                'review': review,
                'rating': rating
            })
    try:
        driver.execute_script(
            "arguments[0].scrollIntoView();",
            driver.find_elements(By.CSS_SELECTOR, 'div.jftiEf')[-1]
        )
    except:
        break
    time.sleep(2)
```

```python
# Menyimpan hasil dari scraping GMAPS
pd.DataFrame(reviews).to_csv("ulasan_uad.csv", index=False)
print(f"[INFO] {len(reviews)} ulasan berhasil didapatkan dan di eksport")
driver.quit
```

# Crawling

```python
from google_play_scraper import search, app
import pandas as pd
import time
```

```python
# --- Konfigurasi Awal ---
KEYWORDS = ["photo editor", "racing game", "weather app", "productivity tool"]
MAX_RESULTS_PER_KEYWORD = 5 # Kita batasi 50 untuk contoh, agar tidak terlalu lama
DETAIL_SLEEP_TIME = 0.5 # Jeda antar pengambilan detail (penting untuk menghindari blokir)

all_detail_data = [] # List untuk menyimpan data detail lengkap

print(f"Memulai proses crawling untuk {len(KEYWORDS)} kata kunci.")
```

```python
# --- Langkah 1: Mencari (Initial Scraping/Gathering IDs) ---
for keyword in KEYWORDS:
    print(f"\n[1/2] Mencari: '{keyword}'...")
    
    # Mencari aplikasi berdasarkan kata kunci (mendapatkan ID dan ringkasan)
    try:
        search_results = search(
            keyword, 
            lang="en", 
            country="us", 
            n_hits=MAX_RESULTS_PER_KEYWORD
        )
    except Exception as e:
        print(f"Error saat mencari '{keyword}': {e}")
        search_results = []
        
    print(f"Ditemukan {len(search_results)} aplikasi.")
```

```python
# --- Langkah 2: Merayap (Crawling/Detail Scraping) ---

for i, r in enumerate(search_results):
    app_id = r['appId']
    print(f"  > [2/2] Mengambil detail app {i+1}/{len(search_results)}: {app_id}")

    try:
        app_detail = app(app_id, lang='en', country='us')
        all_detail_data.append(app_detail)
    except Exception as e:
        print(f"    ! Gagal mengambil detail untuk {app_id}: {e}")

    time.sleep(DETAIL_SLEEP_TIME)
```

```python
# --- Langkah 3: Membuat DataFrame ---
print("\nProses crawling selesai. Membuat DataFrame...")
df_detail = pd.DataFrame(all_detail_data)
```

```python
# --- Langkah 4: Tampilkan Hasil ---
print(f"Total aplikasi dengan detail lengkap: {len(df_detail)}.")
```

```python
# Tampilkan beberapa kolom penting agar rapi
print("\n--- Hasil DataFrame (5 Baris Pertama) ---")
if not df_detail.empty:
    tampilan_rapi = df_detail[['title', 'score', 'genre', 'installs', 'price', 'developer', 'appId']].head()
    print(tampilan_rapi.to_markdown(index=False)) # Menggunakan markdown untuk tabel yang rapi
else:
    print("Tidak ada data yang berhasil dikumpulkan.")
```

```python
# Simpan ke CSV (tanpa menyertakan index angka 0,1,2...)
df_detail.to_csv('hasil_crawling_google_play.csv', index=False)

print("File berhasil disimpan sebagai 'hasil_crawling_google_play.csv'")
```
